```python

model_id = "TheBloke/llama-2-7b-chat-GPTQ"

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto"
)

tokenizer = AutoTokenizer.from_pretrained(model_id)
tokenizer.pad_token = tokenizer.eos_token

print("Model and tokenizer loaded successfully!")
```

```python
dataset = load_dataset("openai/gsm8k", "main")
test_data = dataset["test"]

# Sample a smaller subset
subset = random.sample(list(test_data), 100)
```

```python
base_prompt = """You are a helpful math assistant. Solve each problem step by step.
Always write out your reasoning in detail before giving the final answer.
Conclude each solution by clearly stating: "Therefore, the answer is X."

Q1: There are 15 trees in the grove. Grove workers will plant trees in the grove today. After they are done, there will be 21 trees. How many trees did the grove workers plant today?
A1: Let's think step by step. There are 15 trees originally. Later, there are 21 trees. The number of trees planted is 21 - 15 = 6. Therefore, the answer is 6.

Q2: If there are 3 cars in the parking lot and 2 more cars arrive, how many cars are in the parking lot?
A2: Let's think step by step. There are 3 cars initially. 2 more cars arrive. 3 + 2 = 5. Therefore, the answer is 5.

Q3: Leah had 32 chocolates and her sister had 42. If they ate 35, how many pieces do they have left in total?
A3: Let's think step by step. Total chocolates = 32 + 42 = 74. They ate 35. 74 - 35 = 39. Therefore, the answer is 39.

Q4: Jason had 20 lollipops. He gave Denny some lollipops. Now Jason has 12 lollipops. How many lollipops did Jason give to Denny?
A4: Let's think step by step. Jason started with 20 lollipops. After giving some to Denny, he has 12. So he gave 20 - 12 = 8 to Denny. Therefore, the answer is 8.

Q5: Shawn has 5 toys. For Christmas, he got two toys each from his mom and dad. How many toys does he have now?
A5: Let's think step by step. Shawn started with 5 toys. He got 2 from his mom and 2 from his dad: 5 + 2 + 2 = 9. Therefore, the answer is 9.

Q6: There were nine computers in the server room. Five more computers were installed each day, from Monday to Thursday. How many computers are now in the server room?
A6: Let's think step by step. There are 4 days from Monday to Thursday. Each day, 5 computers were added: 4 × 5 = 20. Initially there were 9 computers. 9 + 20 = 29. Therefore, the answer is 29.

Q7: Michael had 58 golf balls. On Tuesday, he lost 23 golf balls. On Wednesday, he lost 2 more. How many golf balls does he have at the end of Wednesday?
A7: Let's think step by step. Michael started with 58 golf balls. Lost 23 on Tuesday: 58 - 23 = 35. Lost 2 more on Wednesday: 35 - 2 = 33. Therefore, the answer is 33.

Q8: Olivia has $23. She bought five bagels for $3 each. How much money does she have left?
A8: Let's think step by step. She spent 5 × 3 = 15 dollars. She started with $23. 23 - 15 = 8. Therefore, the answer is 8.

Q: {question}
A: Let's think step by step.
"""
```

```python
def generate_multi_step_cot(question, temperatures=[0.3, 0.5, 0.7, 0.9, 1.0]):
    """
    Generates multiple CoT reasoning outputs for a given question
    by varying the temperature parameter for diversity.
    """
    cot_outputs = []

    for temp in temperatures:
        prompt = base_prompt.format(question=question)
        inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

        with torch.no_grad():
            output = model.generate(
                **inputs,
                max_new_tokens=150,
                temperature=temp,
                top_p=0.95,
                do_sample=True,
                pad_token_id=tokenizer.eos_token_id
            )

        decoded = tokenizer.decode(output[0], skip_special_tokens=True)
        answer_segment = decoded.split("Q:")[-1]  # isolate the last Q/A
        cot_outputs.append(answer_segment.strip())

    return cot_outputs
```

```python
def self_reflect_on_cot(cot_text):
    """
    Performs structured self-reflection on a single CoT response.
    The model verifies intermediate steps, identifies errors, and produces a corrected answer.
    """
    reflection_prompt = f"""You are a reasoning verifier.
Review the following reasoning carefully. Identify any calculation or logic mistakes, correct them, and restate the correct reasoning and final answer.

Reasoning:
{cot_text}

Reflection:
"""
    inputs = tokenizer(reflection_prompt, return_tensors="pt").to("cuda")

    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens=150,
            temperature=0.3,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    reflected_text = tokenizer.decode(output[0], skip_special_tokens=True)
    reflection_segment = reflected_text.split("Reflection:")[-1].strip()
    return reflection_segment
```

```python
import re
def majority_voting(reflected_outputs):
    """
    Aggregates all final answers using majority voting on extracted numeric values.
    """
    answers = []
    for out in reflected_outputs:
        nums = re.findall(r"[-+]?\d*\.\d+|\d+", out)
        if nums:
            answers.append(nums[-1])
    if not answers:
        return "No numeric answer found"
    return Counter(answers).most_common(1)[0][0]
```

```python
def run_full_pipeline(question):
    """
    Runs the entire process: multi-step CoT generation, self-reflection, and majority voting.
    """
    print("\n Generating multiple CoT outputs...")
    cot_outputs = generate_multi_step_cot(question)
    print(f"\nGenerated {len(cot_outputs)} reasoning chains.")

    print("\n🔍 Running self-reflection on each chain...")
    reflected = [self_reflect_on_cot(c) for c in cot_outputs]

    for i, r in enumerate(reflected, 1):
        print(f'CoT #{i} \n {cot_outputs[i-1]}')
        print(f"Reflection #{i}\n{r}")

    # print("\n🗳️ Performing majority voting...")
    # final = majority_voting(reflected)

    # print("\n✅ Final Answer (by Majority Vote):", final)
    # return {
    #     "cot_outputs": cot_outputs,
    #     "reflections": reflected,
    #     "final_answer": final
    # }
    return

for sample in subset[0:100]:
    question = sample["question"]
    run_full_pipeline(question)
```

```python
def reflection_check(question, reasoning_output):
    reflection_prompt = f"""
You previously solved the following problem and gave an answer.

Problem: {question}

Your reasoning:
{reasoning_output}

Please check your reasoning carefully.
Follow this structure:
1. Step Verification: Review each step logically and mathematically.
2. Error Identification: Note if any steps or assumptions were incorrect.
3. Corrected Reasoning: Provide the corrected reasoning if needed.
4. Final Answer: State the revised or confirmed final answer.

Your reflection:
"""
    reflection = generate_response(reflection_prompt, temperature=0.4)
    return reflection
```
