```python
import json, random, os

# ====== CONFIG ======
SRC = "/mnt/current/datasets/coco2017/annotations/coco_karpathy_train.json"
DST = "/mnt/current/datasets/coco2017/annotations/coco_karpathy_train_unique.json"
SEED = 42  # for reproducibility
# ====================

random.seed(SEED)

with open(SRC, "r", encoding="utf-8") as f:
    data = json.load(f)

print(f"Loaded {len(data)} samples")

# Some entries have "image_id", others have "image" (filename)
key = "image_id" if "image_id" in data[0] else "image"

# Group all captions by image ID / name
grouped = {}
for entry in data:
    img_id = entry[key]
    grouped.setdefault(img_id, []).append(entry)

print(f"Found {len(grouped)} unique images")

# Pick one random caption per image
unique_entries = [random.choice(entries) for entries in grouped.values()]

# Optional: sort for reproducibility
unique_entries.sort(key=lambda x: x[key])

# Save new JSON
os.makedirs(os.path.dirname(DST), exist_ok=True)
with open(DST, "w", encoding="utf-8") as f:
    json.dump(unique_entries, f, indent=2)

print(f"✅ Saved {len(unique_entries)} entries → {DST}")
```

```python
import os, json, torch
from tqdm import tqdm
from PIL import Image
from torchvision import transforms
from models.blip import blip_decoder

# ========= CONFIG =========
image_root = "/mnt/current/datasets/coco2017/train2014"
ann_path = "/mnt/current/datasets/coco2017/annotations/coco_karpathy_train.json"
output_path = "/mnt/current/datasets/coco2017/annotations/coco_karpathy_train_multi.json"
checkpoint_path = "/mnt/current/BLIP/output/Caption_coco/checkpoint_best.pth"

num_captions = 1      # captions per image
top_p = 0.9           # nucleus sampling
max_len = 20
image_size = 384
vit_type = "base"
# ===========================

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("Loading BLIP caption generator...")
model = blip_decoder(pretrained=checkpoint_path, image_size=image_size, vit=vit_type)
model.eval().to(device)

# same preprocessing used by BLIP
transform = transforms.Compose([
    transforms.Resize((image_size, image_size), interpolation=Image.BICUBIC),
    transforms.ToTensor(),
    transforms.Normalize((0.48145466, 0.4578275, 0.40821073),
                         (0.26862954, 0.26130258, 0.27577711))
])

print(f"Loading annotations from {ann_path} ...")
with open(ann_path, "r") as f:
    data = json.load(f)

# Extract unique images from the COCO Karpathy split
unique_images = {item["image"] for item in data}
print(f"Found {len(unique_images)} unique training images")

results = []

for image_name in tqdm(sorted(unique_images), desc="Generating captions"):
    image_path = os.path.join(image_root, image_name)
    if not os.path.exists(image_path):
        print(f"⚠️ Missing image: {image_path}")
        continue

    try:
        raw_image = Image.open(image_path).convert("RGB")
    except Exception as e:
        print(f"⚠️ Skipping {image_path}: {e}")
        continue

    image_tensor = transform(raw_image).unsqueeze(0).to(device)

    # Generate multiple captions using nucleus sampling
    with torch.no_grad():
        captions = model.generate(
            image_tensor,
            sample=True,
            top_p=top_p,
            num_return_sequences=num_captions,
            max_length=max_len
        )

    for c in captions:
        results.append({"image": image_name, "caption": c})

# Save the dataset
os.makedirs(os.path.dirname(output_path), exist_ok=True)
with open(output_path, "w") as f:
    json.dump(results, f, indent=2)

print(f"\n✅ Done! Generated {len(results)} captions for {len(unique_images)} images.")
print(f"Output saved to: {output_path}")
```

```python
import json

# Define the input and output file names
file1 = '/mnt/current/datasets/coco2017/annotations/coco_karpathy_train_old.json'
file2 = '/mnt/current/datasets/coco2017/annotations/coco_karpathy_train_synthetic_only.json'
output_file = '/mnt/current/datasets/coco2017/annotations/merged.json'

# --- Read and load the first JSON file ---
try:
    with open(file1, 'r') as f:
        data1 = json.load(f)
except FileNotFoundError:
    print(f"Error: {file1} not found. Creating an empty list.")
    data1 = []
except json.JSONDecodeError:
    print(f"Error decoding JSON from {file1}. Please check its format.")
    data1 = []


# --- Read and load the second JSON file ---
try:
    with open(file2, 'r') as f:
        data2 = json.load(f)
except FileNotFoundError:
    print(f"Error: {file2} not found. Creating an empty list.")
    data2 = []
except json.JSONDecodeError:
    print(f"Error decoding JSON from {file2}. Please check its format.")
    data2 = []

# --- Merge the data from both files ---
# The '+' operator concatenates the two lists
merged_data = data1 + data2

# --- Write the merged data to the output file ---
with open(output_file, 'w') as f:
    # Use indent=2 for pretty-printing the JSON output
    json.dump(merged_data, f, indent=2)

print(f"✅ Successfully merged {file1} and {file2} into {output_file}.")
print(f"Total entries in merged file: {len(merged_data)}")
```

```python
import json

# Define the input and output file names
file1 = '/mnt/current/datasets/coco2017/annotations/coco_karpathy_train_old.json'
file2 = '/mnt/current/datasets/coco2017/annotations/coco_karpathy_train_synthetic_only.json'
output_file = '/mnt/current/datasets/coco2017/annotations/merged_50_50.json'

# --- Read and load the first JSON file ---
try:
    with open(file1, 'r') as f:
        data1 = json.load(f)
except FileNotFoundError:
    print(f"Error: {file1} not found. Creating an empty list.")
    data1 = []
except json.JSONDecodeError:
    print(f"Error decoding JSON from {file1}. Please check its format.")
    data1 = []


# --- Read and load the second JSON file ---
try:
    with open(file2, 'r') as f:
        data2 = json.load(f)
except FileNotFoundError:
    print(f"Error: {file2} not found. Creating an empty list.")
    data2 = []
except json.JSONDecodeError:
    print(f"Error decoding JSON from {file2}. Please check its format.")
    data2 = []

# --- Calculate 50% of each dataset ---
# Using integer division '//' to get half the length of each list
half_data1 = len(data1) // 2
half_data2 = len(data2) // 2

# --- Take the first 50% from each list and merge them ---
# Slicing the lists to get the first half of the entries
merged_data = data1[:half_data1] + data2[:half_data2]

# --- Write the merged data to the output file ---
with open(output_file, 'w') as f:
    # Use indent=2 for pretty-printing the JSON output
    json.dump(merged_data, f, indent=2)

print(f"✅ Successfully merged 50% of {file1} and 50% of {file2} into {output_file}.")
print(f"Total entries in merged file: {len(merged_data)}")
print(f"Entries from {file1}: {half_data1}")
print(f"Entries from {file2}: {half_data2}")
```

## image extraction for paper

```python
import os
import json
import random
import shutil
from zipfile import ZipFile
from pathlib import Path

# ---------------- CONFIG ----------------
IMAGE_ROOT = "/mnt/current/datasets/coco2017/val2014"  # image folder
HUMAN_JSON = "/mnt/current/datasets/coco2017/annotations/coco_karpathy_test.json"
GEN_JSON = "/mnt/current/BLIP/output/Caption_coco/result/test_epoch0.json"
NUM_IMAGES = 10
OUTPUT_ZIP = "/mnt/current/qualitative_sample.zip"
RANDOM_SEED = 42
# ----------------------------------------

random.seed(RANDOM_SEED)

# Create temp folder to collect files
TEMP_DIR = Path("temp_qualitative")
if TEMP_DIR.exists():
    shutil.rmtree(TEMP_DIR)
TEMP_DIR.mkdir(parents=True, exist_ok=True)

# Load human annotations
with open(HUMAN_JSON, "r") as f:
    human_data = json.load(f)

human_dict = {}
# Supports list of dicts: {"image":..., "caption":...} or {"image_id":..., "caption":...}
if "annotations" in human_data and "images" in human_data:
    # COCO-style
    id_to_file = {img["id"]: img["file_name"] for img in human_data["images"]}
    for ann in human_data["annotations"]:
        fname = id_to_file.get(ann["image_id"])
        if fname not in human_dict:
            human_dict[fname] = []
        human_dict[fname].append(ann["caption"])
else:
    # Simple list format
    for item in human_data:
        key = item.get("image") or str(item.get("image_id"))
        if key not in human_dict:
            human_dict[key] = []
        human_dict[key].append(item["caption"])

# Load generated captions
with open(GEN_JSON, "r") as f:
    gen_data = json.load(f)

gen_dict = {}
for item in gen_data:
    key = item.get("image") or str(item.get("image_id"))
    if key not in gen_dict:
        gen_dict[key] = []
    gen_dict[key].append(item["caption"])

# Get common images
common_images = list(set(human_dict.keys()) & set(gen_dict.keys()))
if len(common_images) < NUM_IMAGES:
    raise ValueError(f"Not enough common images ({len(common_images)}) to sample {NUM_IMAGES}")

sample_images = random.sample(common_images, NUM_IMAGES)

# Create folders for selection
IMG_DIR = TEMP_DIR / "images"
HUMAN_DIR = TEMP_DIR / "human_annotations"
GEN_DIR = TEMP_DIR / "generated_annotations"
IMG_DIR.mkdir()
HUMAN_DIR.mkdir()
GEN_DIR.mkdir()

for fname in sample_images:
    # Copy image file
    src_img_path = os.path.join(IMAGE_ROOT, fname)
    if not os.path.exists(src_img_path):
        print(f"⚠️ Image not found: {src_img_path}, skipping")
        continue
    shutil.copy(src_img_path, IMG_DIR / fname)
    
    # Save human annotation
    human_file = HUMAN_DIR / f"{fname}.txt"
    with open(human_file, "w", encoding="utf-8") as f:
        for cap in human_dict[fname]:
            f.write(cap + "\n")
    
    # Save generated caption
    gen_file = GEN_DIR / f"{fname}.txt"
    with open(gen_file, "w", encoding="utf-8") as f:
        for cap in gen_dict[fname]:
            f.write(cap + "\n")

# Create zip file
with ZipFile(OUTPUT_ZIP, "w") as zipf:
    for folder in [IMG_DIR, HUMAN_DIR, GEN_DIR]:
        for file in folder.glob("*"):
            zipf.write(file, arcname=os.path.join(folder.name, file.name))

# Clean up temp folder
shutil.rmtree(TEMP_DIR)

print(f"✅ Done! Created {OUTPUT_ZIP} with {NUM_IMAGES} samples.")
```
