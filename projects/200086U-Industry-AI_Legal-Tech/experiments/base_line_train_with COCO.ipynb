```python
import json, random, os

# ====== CONFIG ======
SRC = "/mnt/current/datasets/coco2017/annotations/coco_karpathy_train.json"
DST = "/mnt/current/datasets/coco2017/annotations/coco_karpathy_train_unique.json"
SEED = 42  # for reproducibility
# ====================

random.seed(SEED)

with open(SRC, "r", encoding="utf-8") as f:
    data = json.load(f)

print(f"Loaded {len(data)} samples")

# Some entries have "image_id", others have "image" (filename)
key = "image_id" if "image_id" in data[0] else "image"

# Group all captions by image ID / name
grouped = {}
for entry in data:
    img_id = entry[key]
    grouped.setdefault(img_id, []).append(entry)

print(f"Found {len(grouped)} unique images")

# Pick one random caption per image
unique_entries = [random.choice(entries) for entries in grouped.values()]

# Optional: sort for reproducibility
unique_entries.sort(key=lambda x: x[key])

# Save new JSON
os.makedirs(os.path.dirname(DST), exist_ok=True)
with open(DST, "w", encoding="utf-8") as f:
    json.dump(unique_entries, f, indent=2)

print(f"✅ Saved {len(unique_entries)} entries → {DST}")
```

```python
import json, os

SRC = "/mnt/current/datasets/coco2017/annotations/coco_gt/coco_karpathy_val_gt.json"
DST = "/mnt/current/datasets/coco2017/annotations/coco_gt/coco_karpathy_val_gt_fixed.json"

with open(SRC, "r", encoding="utf-8") as f:
    gt = json.load(f)

# If the file is a dict with "annotations" only, fix it
if isinstance(gt, dict) and "annotations" in gt:
    anns = gt["annotations"]
else:
    anns = gt  # if it's a plain list

# Build image list
image_ids = sorted(set(a["image_id"] for a in anns))
images = [{"id": i, "file_name": f"COCO_train2014_{i:012d}.jpg"} for i in image_ids]

fixed = {
    "info": {
        "description": "COCO Caption Karpathy Split (Fixed)",
        "version": "1.0",
        "year": 2014,
        "contributor": "User-generated fix",
        "date_created": "2025-10-04"
    },
    "licenses": [
        {
            "id": 1,
            "name": "Attribution-NonCommercial-ShareAlike License",
            "url": "http://creativecommons.org/licenses/by-nc-sa/2.0/"
        }
    ],
    "images": images,
    "annotations": anns
}

os.makedirs(os.path.dirname(DST), exist_ok=True)
with open(DST, "w", encoding="utf-8") as f:
    json.dump(fixed, f, indent=2)

print(f"✅ Fixed GT file saved → {DST}")
```

```python
import json, random

# === CONFIG ===
input_json = "/mnt/current/datasets/coco2017/annotations/coco_karpathy_train.json"
output_json = "/mnt/current/datasets/coco2017/annotations/coco_karpathy_train_2pct.json"
subset_percent = 1
seed = 42
# ==============

random.seed(seed)

# Load data
with open(input_json, "r") as f:
    data = json.load(f)

total = len(data)
subset_size = max(1, int(total * subset_percent / 100))
print(f"Loaded {total} entries. Selecting {subset_size} ({subset_percent}%)...")

# Random sample
subset = random.sample(data, subset_size)

# Save subset
with open(output_json, "w") as f:
    json.dump(subset, f, indent=2)

print(f"✅ Saved {len(subset)} entries → {output_json}")
```
