# GenCast Mini Demo

This notebook demonstrates running `GenCast 1p0deg Mini <2019`.

`GenCast 1p0deg Mini <2019` is a GenCast model at 1deg resolution, with 13 pressure levels and a 4 times refined icosahedral mesh. It is trained on ERA5 data from 1979 to 2018, and can be causally evaluated on 2019 and later years.

While other GenCast models are [available](https://github.com/google-deepmind/graphcast/blob/main/README.md), this model has the smallest memory footprint of those provided and is the only one runnable with the freely provided TPUv2-8 configuration in Colab. You can select this configuration in `Runtime>Change Runtime Type`.

**N.B.** The performance of `GenCast 1p0deg Mini <2019` is reasonable but is not representative of the performance of the other GenCast models described in the [README](https://github.com/google-deepmind/graphcast/blob/main/README.md).

To run the other models using Google Cloud Compute, refer to [gencast_demo_cloud_vm.ipynb](https://colab.research.google.com/github/deepmind/graphcast/blob/master/gencast_demo_cloud_vm.ipynb).

```python
# @title Plotting functions

def select(
    data: xarray.Dataset,
    variable: str,
    level: Optional[int] = None,
    max_steps: Optional[int] = None
    ) -> xarray.Dataset:
  data = data[variable]
  if "batch" in data.dims:
    data = data.isel(batch=0)
  if max_steps is not None and "time" in data.sizes and max_steps < data.sizes["time"]:
    data = data.isel(time=range(0, max_steps))
  if level is not None and "level" in data.coords:
    data = data.sel(level=level)
  return data

def scale(
    data: xarray.Dataset,
    center: Optional[float] = None,
    robust: bool = False,
    ) -> tuple[xarray.Dataset, matplotlib.colors.Normalize, str]:
  vmin = np.nanpercentile(data, (2 if robust else 0))
  vmax = np.nanpercentile(data, (98 if robust else 100))
  if center is not None:
    diff = max(vmax - center, center - vmin)
    vmin = center - diff
    vmax = center + diff
  return (data, matplotlib.colors.Normalize(vmin, vmax),
          ("RdBu_r" if center is not None else "viridis"))

def plot_data(
    data: dict[str, xarray.Dataset],
    fig_title: str,
    plot_size: float = 5,
    robust: bool = False,
    cols: int = 4
    ) -> tuple[xarray.Dataset, matplotlib.colors.Normalize, str]:

  first_data = next(iter(data.values()))[0]
  max_steps = first_data.sizes.get("time", 1)
  assert all(max_steps == d.sizes.get("time", 1) for d, _, _ in data.values())

  cols = min(cols, len(data))
  rows = math.ceil(len(data) / cols)
  figure = plt.figure(figsize=(plot_size * 2 * cols,
                               plot_size * rows))
  figure.suptitle(fig_title, fontsize=16)
  figure.subplots_adjust(wspace=0, hspace=0)
  figure.tight_layout()

  images = []
  for i, (title, (plot_data, norm, cmap)) in enumerate(data.items()):
    ax = figure.add_subplot(rows, cols, i+1)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(title)
    im = ax.imshow(
        plot_data.isel(time=0, missing_dims="ignore"), norm=norm,
        origin="lower", cmap=cmap)
    plt.colorbar(
        mappable=im,
        ax=ax,
        orientation="vertical",
        pad=0.02,
        aspect=16,
        shrink=0.75,
        cmap=cmap,
        extend=("both" if robust else "neither"))
    images.append(im)

  def update(frame):
    if "time" in first_data.dims:
      td = datetime.timedelta(microseconds=first_data["time"][frame].item() / 1000)
      figure.suptitle(f"{fig_title}, {td}", fontsize=16)
    else:
      figure.suptitle(fig_title, fontsize=16)
    for im, (plot_data, norm, cmap) in zip(images, data.values()):
      im.set_data(plot_data.isel(time=frame, missing_dims="ignore"))

  ani = animation.FuncAnimation(
      fig=figure, func=update, frames=max_steps, interval=250)
  plt.close(figure.number)
  return HTML(ani.to_jshtml())
```

Choose one of the two ways of getting model params:
- **checkpoint**: You'll get sensible predictions, but are limited to the model architecture that it was trained with, which may not fit on your device.
- **random**: You'll get random predictions, but you can change the model architecture and data resolution which may run faster or fit on your device.



```python
# @title Choose the model

params_file_options = [
    name for blob in gcs_bucket.list_blobs(prefix=(dir_prefix+"params/"))
    if (name := blob.name.removeprefix(dir_prefix+"params/"))]  # Drop empty string.

latent_value_options = [int(2**i) for i in range(4, 10)]
random_latent_size = widgets.Dropdown(
    options=latent_value_options, value=512,description="Latent size:")
random_attention_type = widgets.Dropdown(
    options=["splash_mha", "triblockdiag_mha", "mha"], value="splash_mha", description="Attention:")
random_mesh_size = widgets.IntSlider(
    value=4, min=4, max=6, description="Mesh size:")
random_num_heads = widgets.Dropdown(
    options=[int(2**i) for i in range(0, 3)], value=4,description="Num heads:")
random_attention_k_hop = widgets.Dropdown(
    options=[int(2**i) for i in range(2, 5)], value=16,description="Attn k hop:")
random_resolution = widgets.Dropdown(
    options=["1p0", "0p25"], value="1p0", description="Resolution:")

def update_latent_options(*args):
  def _latent_valid_for_attn(attn, latent, heads):
    head_dim, rem = divmod(latent, heads)
    if rem != 0:
      return False
    # Required for splash attn.
    if head_dim % 128 != 0:
      return attn != "splash_mha"
    return True
  attn = random_attention_type.value
  heads = random_num_heads.value
  random_latent_size.options = [
      latent for latent in latent_value_options
      if _latent_valid_for_attn(attn, latent, heads)]

# Observe changes to only allow for valid combinations.
random_attention_type.observe(update_latent_options, "value")
random_latent_size.observe(update_latent_options, "value")
random_num_heads.observe(update_latent_options, "value")

params_file = widgets.Dropdown(
    options=[f for f in params_file_options if "Mini" in f],
    description="Params file:",
    layout={"width": "max-content"})

source_tab = widgets.Tab([
    params_file,
    widgets.VBox([
        random_attention_type,
        random_mesh_size,
        random_num_heads,
        random_latent_size,
        random_attention_k_hop,
        random_resolution
    ]),
])
source_tab.set_title(0, "Checkpoint")
source_tab.set_title(1, "Random")
widgets.VBox([
    source_tab,
    widgets.Label(value="Run the next cell to load the model. Rerunning this cell clears your selection.")
])
```

```python
# @title Load the model

source = source_tab.get_title(source_tab.selected_index)

if source == "Random":
  params = None  # Filled in below
  state = {}
  task_config = gencast.TASK
  # Use default values.
  sampler_config = gencast.SamplerConfig()
  noise_config = gencast.NoiseConfig()
  noise_encoder_config = denoiser.NoiseEncoderConfig()
  # Configure, otherwise use default values.
  denoiser_architecture_config = denoiser.DenoiserArchitectureConfig(
    sparse_transformer_config = denoiser.SparseTransformerConfig(
        attention_k_hop=random_attention_k_hop.value,
        attention_type=random_attention_type.value,
        d_model=random_latent_size.value,
        num_heads=random_num_heads.value
        ),
    mesh_size=random_mesh_size.value,
    latent_size=random_latent_size.value,
  )
else:
  assert source == "Checkpoint"
  with gcs_bucket.blob(dir_prefix + f"params/{params_file.value}").open("rb") as f:
    ckpt = checkpoint.load(f, gencast.CheckPoint)
  params = ckpt.params
  state = {}

  task_config = ckpt.task_config
  sampler_config = ckpt.sampler_config
  noise_config = ckpt.noise_config
  noise_encoder_config = ckpt.noise_encoder_config
  denoiser_architecture_config = ckpt.denoiser_architecture_config
  print("Model description:\n", ckpt.description, "\n")
  print("Model license:\n", ckpt.license, "\n")
```

```python
# @title Build jitted functions, and possibly initialize random weights


def construct_wrapped_gencast():
  """Constructs and wraps the GenCast Predictor."""
  predictor = gencast.GenCast(
      sampler_config=sampler_config,
      task_config=task_config,
      denoiser_architecture_config=denoiser_architecture_config,
      noise_config=noise_config,
      noise_encoder_config=noise_encoder_config,
  )

  predictor = normalization.InputsAndResiduals(
      predictor,
      diffs_stddev_by_level=diffs_stddev_by_level,
      mean_by_level=mean_by_level,
      stddev_by_level=stddev_by_level,
  )

  predictor = nan_cleaning.NaNCleaner(
      predictor=predictor,
      reintroduce_nans=True,
      fill_value=min_by_level,
      var_to_clean='sea_surface_temperature',
  )

  return predictor


@hk.transform_with_state
def run_forward(inputs, targets_template, forcings):
  predictor = construct_wrapped_gencast()
  return predictor(inputs, targets_template=targets_template, forcings=forcings)


@hk.transform_with_state
def loss_fn(inputs, targets, forcings):
  predictor = construct_wrapped_gencast()
  loss, diagnostics = predictor.loss(inputs, targets, forcings)
  return xarray_tree.map_structure(
      lambda x: xarray_jax.unwrap_data(x.mean(), require_jax=True),
      (loss, diagnostics),
  )


def grads_fn(params, state, inputs, targets, forcings):
  def _aux(params, state, i, t, f):
    (loss, diagnostics), next_state = loss_fn.apply(
        params, state, jax.random.PRNGKey(0), i, t, f
    )
    return loss, (diagnostics, next_state)

  (loss, (diagnostics, next_state)), grads = jax.value_and_grad(
      _aux, has_aux=True
  )(params, state, inputs, targets, forcings)
  return loss, diagnostics, next_state, grads


if params is None:
  init_jitted = jax.jit(loss_fn.init)
  params, state = init_jitted(
      rng=jax.random.PRNGKey(0),
      inputs=train_inputs,
      targets=train_targets,
      forcings=train_forcings,
  )


loss_fn_jitted = jax.jit(
    lambda rng, i, t, f: loss_fn.apply(params, state, rng, i, t, f)[0]
)
grads_fn_jitted = jax.jit(grads_fn)
run_forward_jitted = jax.jit(
    lambda rng, i, t, f: run_forward.apply(params, state, rng, i, t, f)[0]
)
# We also produce a pmapped version for running in parallel.
run_forward_pmap = xarray_jax.pmap(run_forward_jitted, dim="sample")
```

# Run the model

The `chunked_prediction_generator_multiple_runs` iterates over forecast steps, where the 1 step forecast is jitted and samples are pmapped across the chips.
This allows us to make efficient use of all devices and parallelise generating an ensemble across them. We then combine the chunks at the end to form our final forecast.

Note that the `Autoregressive rollout` cell will take longer than the standard inference time to run when executed for the first time, as this will include code compilation time. This cost does not increase with the number of devices, it is a fixed-cost one time operation whose result can be reused across any number of devices.

```python
import numpy as np
import xarray as xr
import scipy.ndimage as nd

# --------------------------
# Helpers (same ideas as v3)
# --------------------------
def _sigma_for_array(arr: xr.DataArray):
    """Return gaussian_filter sigma vector (keeps same smoothing style as mid-eval)."""
    ndims = len(arr.dims)
    sigma = [0] * ndims
    for i, d in enumerate(arr.dims):
        if d in ("lat", "lon"):
            sigma[i] = 1
        elif d == "time":
            sigma[i] = 0  # keep time smoothing minimal/none by default
    return sigma

def _divergence_safe(u_da: xr.DataArray, v_da: xr.DataArray) -> xr.DataArray:
    """
    Approx horizontal divergence using finite differences.
    Returns DataArray same dims/coords as inputs.
    This is intentionally simple/robust (not high-precision).
    """
    # Require lat & lon dims to exist
    if ("lat" not in u_da.dims) or ("lon" not in u_da.dims):
        # fallback zero field
        return xr.zeros_like(u_da)
    # axis indices
    lat_ax = list(u_da.dims).index("lat")
    lon_ax = list(u_da.dims).index("lon")
    lat_vals = u_da.coords["lat"].values
    lon_vals = u_da.coords["lon"].values

    # compute gradients with np.gradient using angular spacing
    # np.gradient needs spacing arrays for lat/lon in radians
    lat_rad = np.deg2rad(lat_vals)
    lon_rad = np.deg2rad(lon_vals)
    grads_u = np.gradient(u_da.values, np.deg2rad(lat_vals), np.deg2rad(lon_vals), axis=(lat_ax, lon_ax))
    grads_v = np.gradient(v_da.values, np.deg2rad(lat_vals), np.deg2rad(lon_vals), axis=(lat_ax, lon_ax))
    dudphi = grads_u[0]; dudlam = grads_u[1]
    dvdphi = grads_v[0]; dvdlam = grads_v[1]

    R = 6371000.0
    cosphi = np.cos(np.deg2rad(lat_vals))
    shape = [1] * u_da.ndim
    shape[lat_ax] = len(lat_vals)
    cosphi_b = cosphi.reshape(shape)

    dudx = dudlam / (R * cosphi_b)
    dvdy = dvdphi / R
    div = dudx + dvdy
    return xr.DataArray(div, dims=u_da.dims, coords=u_da.coords)

def sanity_check(orig: xr.Dataset, noisy: xr.Dataset):
    """Raise AssertionError if any var/dim/shape differ."""
    if set(orig.data_vars) != set(noisy.data_vars):
        raise AssertionError("Data variables differ between original and noisy dataset")
    for v in orig.data_vars:
        if list(orig[v].dims) != list(noisy[v].dims):
            raise AssertionError(f"Dim mismatch for {v}: {orig[v].dims} vs {noisy[v].dims}")
        if orig[v].shape != noisy[v].shape:
            raise AssertionError(f"Shape mismatch for {v}: {orig[v].shape} vs {noisy[v].shape}")
    return True
```

```python

# --------------------------
# v4: physically-normalized perturbations (shape-safe)
# --------------------------
def add_custom_noise_v4_safe(
    ds: xr.Dataset,
    *,
    rel_strength: float = 0.001,       # base relative fraction for multiplicative vars
    abs_strength: float = 0.01,        # base absolute fraction for additive vars (fraction of std)
    var_scale: dict | None = None,
    tropics_boost: float = 0.5,
    wind_to_z_alpha: float = 0.0,      # keep 0.0 by default; enable if you know what you're doing
    z_to_t_beta: float = 0.0,
    seed: int | None = None,
    apply_to_vars: list | None = None,
    max_rel_frac: float = 0.02,        # max allowed relative perturbation (e.g. 2%) for multiplicative vars
    temporal_decorrelation: bool = True,
) -> xr.Dataset:
    """
    v4: safer perturbations that:
      - use multiplicative (relative) noise for large-magnitude fields (geopotential, pressure, temp)
      - use additive/smoothed noise for winds, humidity, vertical velocity
      - cap relative noise to max_rel_frac
      - preserve dataset shapes/dims exactly (in-place .values writes)
      - optional (disabled by default) wind->z coupling
    """

    if seed is not None:
        np.random.seed(seed)
    if var_scale is None:
        var_scale = {}

    # default list similar to earlier versions
    if apply_to_vars is None:
        apply_to_vars = [v for v in [
            'u_component_of_wind','v_component_of_wind',
            'temperature','vertical_velocity',
            'specific_humidity','2m_temperature',
            '10m_u_component_of_wind','10m_v_component_of_wind',
            'sea_surface_temperature'
        ] if v in ds.data_vars]

    noisy_ds = ds.copy(deep=True)
    lat_mask = (abs(ds['lat']) < 30).astype(float)

    # Define which variables we treat multiplicatively (relative) vs additively
    multiplicative_vars = set([

        'temperature',            # 3D temperature
        '2m_temperature',
        'sea_surface_temperature'
    ])
    additive_vars = set(apply_to_vars) - multiplicative_vars

    # ---------------- Wind noise (additive) ----------------
    u_name = 'u_component_of_wind' if 'u_component_of_wind' in ds.data_vars else None
    v_name = 'v_component_of_wind' if 'v_component_of_wind' in ds.data_vars else None
    wind_noise_u = wind_noise_v = None

    if u_name and v_name and (u_name in apply_to_vars) and (v_name in apply_to_vars):
        u_arr = ds[u_name]; v_arr = ds[v_name]
        # absolute scale for winds: use std * abs_strength * var_scale
        s_u = float(np.nanstd(u_arr.values)) or 1.0
        s_v = float(np.nanstd(v_arr.values)) or 1.0
        sigma_u_abs = s_u * abs_strength * var_scale.get(u_name, 1.0)
        sigma_v_abs = s_v * abs_strength * var_scale.get(v_name, 1.0)

        noise_u = np.random.normal(size=u_arr.shape).astype(u_arr.dtype)
        noise_v = np.random.normal(size=v_arr.shape).astype(v_arr.dtype)

        if temporal_decorrelation and 'time' in u_arr.dims:
            t_idx = list(u_arr.dims).index('time')
            t_mult = np.random.normal(loc=1.0, scale=0.01, size=(u_arr.sizes['time'],)).astype(u_arr.dtype)
            shape = [1] * noise_u.ndim; shape[t_idx] = u_arr.sizes['time']
            noise_u = noise_u * t_mult.reshape(shape)
            noise_v = noise_v * t_mult.reshape(shape)

        smooth_u = nd.gaussian_filter(noise_u, sigma=_sigma_for_array(u_arr))
        smooth_v = nd.gaussian_filter(noise_v, sigma=_sigma_for_array(v_arr))
        smooth_u_da = xr.DataArray(smooth_u, dims=u_arr.dims, coords=u_arr.coords)
        smooth_v_da = xr.DataArray(smooth_v, dims=v_arr.dims, coords=v_arr.coords)
        mask_u = lat_mask.broadcast_like(u_arr)
        wind_noise_u = smooth_u_da * sigma_u_abs * (1.0 + tropics_boost * mask_u)
        wind_noise_v = smooth_v_da * sigma_v_abs * (1.0 + tropics_boost * mask_u)

        noisy_ds[u_name].values = (u_arr + wind_noise_u).values
        noisy_ds[v_name].values = (v_arr + wind_noise_v).values

    # ---------------- Geopotential (and optional coupling) ----------------
    # If wind->z coupling desired, compute divergence of wind_noise and convert
    if wind_noise_u is not None and wind_noise_v is not None and wind_to_z_alpha != 0.0:
        div = _divergence_safe(wind_noise_u, wind_noise_v)
        z_delta = div * wind_to_z_alpha
    else:
        z_delta = None

    # handle multiplicative variables (including geopotential, pressure, temp)
    for var in (multiplicative_vars & set(apply_to_vars)):
        arr = ds[var]
        # compute a robust baseline amplitude using mean absolute value (avoid huge std-driven values).
        # If arr can be negative (e.g., geopotential), use mean absolute.
        mean_abs = float(np.nanmean(np.abs(arr.values)))
        if mean_abs == 0 or np.isnan(mean_abs):
            mean_abs = max(1.0, float(np.nanstd(arr.values)))  # fallback

        # multiplicative noise: relative perturbation sampled from N(0,1) then scaled by rel_strength
        noise_rel = np.random.normal(size=arr.shape).astype(arr.dtype)
        if temporal_decorrelation and 'time' in arr.dims:
            t_idx = list(arr.dims).index('time')
            t_mult = np.random.normal(loc=1.0, scale=0.01, size=(arr.sizes['time'],)).astype(arr.dtype)
            shape = [1] * noise_rel.ndim; shape[t_idx] = arr.sizes['time']
            noise_rel = noise_rel * t_mult.reshape(shape)

        smooth_rel = nd.gaussian_filter(noise_rel, sigma=_sigma_for_array(arr))
        smooth_rel_da = xr.DataArray(smooth_rel, dims=arr.dims, coords=arr.coords)

        # scale relative noise by rel_strength and per-var multiplier
        rel_mult = rel_strength * var_scale.get(var, 1.0)

        # cap relative perturbation to avoid huge jumps (max_rel_frac)
        perturb_rel = smooth_rel_da * rel_mult
        perturb_rel = perturb_rel.where(np.abs(perturb_rel) <= max_rel_frac,
                                        np.sign(perturb_rel) * max_rel_frac)

        # If geopotential and we have z_delta from divergence, combine (small weight)
        if var == 'geopotential' and z_delta is not None:
            # safe expand z_delta to geopotential dims if needed
            z_arr = arr
            if ('level' in z_arr.dims) and (z_delta is not None) and ('level' not in z_delta.dims):
                # expand along level axis
                expand_axis = list(z_arr.dims).index('level')
                z_vals = np.repeat(np.expand_dims(z_delta.values, axis=expand_axis),
                                   z_arr.sizes['level'], axis=expand_axis)
                z_delta_b = xr.DataArray(z_vals, dims=z_arr.dims, coords=z_arr.coords)
            else:
                # attempt safe broadcast
                try:
                    z_delta_b = xr.DataArray(np.broadcast_to(z_delta.values, z_arr.shape),
                                             dims=z_arr.dims, coords=z_arr.coords)
                except Exception:
                    z_delta_b = z_delta.broadcast_like(z_arr)

            # convert divergence-derived absolute perturbation to a small relative fraction
            # relative_from_delta = z_delta_b / mean_abs (small)
            relative_from_delta = (z_delta_b / mean_abs) * 0.001  # tiny contribution; tuneable
            perturb_rel = perturb_rel + relative_from_delta

        # apply multiplicative update: arr * (1 + perturb_rel)
        noisy_ds[var].values = (arr * (1.0 + perturb_rel)).values

    # ---------------- Additive small noise for the rest ----------------
    for var in (additive_vars & set(apply_to_vars)):
        # winds were handled earlier; skip them here
        if var in (u_name, v_name):
            continue
        arr = ds[var]
        s = float(np.nanstd(arr.values))
        if np.isnan(s) or s == 0.0:
            s = 1.0
        sigma_abs = s * abs_strength * var_scale.get(var, 1.0)

        noise = np.random.normal(size=arr.shape).astype(arr.dtype)
        if temporal_decorrelation and 'time' in arr.dims:
            t_idx = list(arr.dims).index('time')
            t_mult = np.random.normal(loc=1.0, scale=0.01, size=(arr.sizes['time'],)).astype(arr.dtype)
            shape = [1] * noise.ndim; shape[t_idx] = arr.sizes['time']
            noise = noise * t_mult.reshape(shape)

        smooth = nd.gaussian_filter(noise, sigma=_sigma_for_array(arr))
        smooth_da = xr.DataArray(smooth, dims=arr.dims, coords=arr.coords)
        mask_for_var = lat_mask.broadcast_like(arr)
        perturb = smooth_da * sigma_abs * (1.0 + tropics_boost * mask_for_var)

        # For some additive vars (e.g., humidity) we may want to enforce bounds (0..1)
        if var in ('specific_humidity',):
            new_vals = arr.values + perturb.values
            # clip physically
            noisy_ds[var].values = np.clip(new_vals, 0.0, None)
        else:
            noisy_ds[var].values = (arr + perturb).values

    # ---------------- Optional: small temp coupling from geopotential (if z_to_t_beta set) ----------------
    # if z_to_t_beta != 0.0 and 'geopotential' in ds.data_vars and 'temperature' in ds.data_vars:
    #     # compute small impact: (z change) -> small fractional temperature change
    #     z_old = ds['geopotential']
    #     z_new = noisy_ds['geopotential']
    #     # compute delta relative
    #     delta_z = (z_new - z_old)
    #     # expand to temperature dims (if necessary) safely
    #     t_arr = ds['temperature']
    #     if ('level' in t_arr.dims) and ('level' not in delta_z.dims):
    #         expand_axis = list(t_arr.dims).index('level')
    #         delta_vals = np.repeat(np.expand_dims(delta_z.values, axis=expand_axis),
    #                                t_arr.sizes['level'], axis=expand_axis)
    #         delta_z_b = xr.DataArray(delta_vals, dims=t_arr.dims, coords=t_arr.coords)
    #     else:
    #         try:
    #             delta_z_b = xr.DataArray(np.broadcast_to(delta_z.values, t_arr.shape),
    #                                      dims=t_arr.dims, coords=t_arr.coords)
    #         except Exception:
    #             delta_z_b = delta_z.broadcast_like(t_arr)

    #     # temperature perturbation is tiny fraction of delta_z
    #     t_perturb = delta_z_b * z_to_t_beta
    #     noisy_ds['temperature'].values = (t_arr + t_perturb).values

    # ---------------- final sanity check ----------------
    sanity_check(ds, noisy_ds)
    return noisy_ds
```

```python


# Your existing CRPS function is fine; keep it as-is:
# def crps(targets, predictions, bias_corrected=True): ...

def mean_crps(preds: xarray.Dataset, targets: xarray.Dataset) -> float:
    """
    Domain- and variable-averaged CRPS (scalar).
    Works whether CRPS returns a Dataset (per-var) or a DataArray.
    """
    crps_map = crps(targets, preds)             # usually an xarray.Dataset
    if isinstance(crps_map, xarray.Dataset):
        crps_arr = crps_map.to_array("variable")  # stack vars -> DataArray
    else:
        crps_arr = crps_map                      # already a DataArray
    return float(crps_arr.mean().item())         # average over all dims

def mean_crps_by_var(preds: xarray.Dataset, targets: xarray.Dataset) -> dict:
    """
    Per-variable CRPS, averaged over space/time/levels (returns dict).
    """
    crps_map = crps(targets, preds)             # Dataset
    assert isinstance(crps_map, xarray.Dataset), "Expected CRPS to be a Dataset"
    out = {}
    for v in crps_map.data_vars:
        out[v] = float(crps_map[v].mean().item())
    return out

def mean_crps_by_lead(preds: xarray.Dataset, targets: xarray.Dataset) -> xarray.DataArray:
    """
    CRPS averaged over space/levels/vars but kept by 'time' (lead-time curve).
    Returns a 1D DataArray over 'time'.
    """
    crps_map = crps(targets, preds)             # Dataset
    crps_arr = crps_map.to_array("variable")    # DataArray[..., time, ...]
    # reduce everything except 'time'
    reduce_dims = [d for d in crps_arr.dims if d != "time"]
    return crps_arr.mean(dim=reduce_dims)
```
