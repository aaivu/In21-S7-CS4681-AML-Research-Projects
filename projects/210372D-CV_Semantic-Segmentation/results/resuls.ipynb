```python
import os
import json
import pandas as pd

base_dir = "."

folders = ["01", "02", "03", "04", "05", "06", "07"]

data = {"epoch": list(range(31))}

for folder in folders:
    result_path = os.path.join(base_dir, folder, "output", "results.json")
    if not os.path.exists(result_path):
        continue

    with open(result_path, "r") as f:
        results = json.load(f)

    data[folder] = results["0"][:31]

df = pd.DataFrame(data)

# df.to_csv("all_results.csv", index=False)

df
```

```python
selected_epochs = [0, 5, 10, 15, 20, 25, 30]

df_selected = df[df["epoch"].isin(selected_epochs)].reset_index(drop=True)

df_selected.to_csv("results.csv", index=False)

df_selected
```

```python
import json
import matplotlib.pyplot as plt

with open("experiment_config.json", "r") as f:
    exp_names = json.load(f)

baseline_values = df[df["epoch"] == 0].iloc[0, 1:]

plot_df = df[df["epoch"] != 0]

for col in plot_df.columns[1:]:
    label = exp_names.get(col, col)
    plt.plot(plot_df["epoch"], plot_df[col], label=label)

# for col, base_val in baseline_values.items():
#     plt.axhline(y=base_val, linestyle="--", color="gray", linewidth=1)

plt.xlabel("Epoch")
plt.ylabel("mIoU")
plt.title("Model Results Across Epochs")
plt.legend()
plt.grid(True)
plt.show()
```
